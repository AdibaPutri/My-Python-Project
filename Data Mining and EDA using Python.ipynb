# HANDLING OUTLIERS

```python
dfc_pay.info()
```

```python
#Handling Outliers
q1 = dfc_pay['payment_value'].quantile(.25)
q3 = dfc_pay['payment_value'].quantile(.75)
iqr = q3-q1

UIF = q3 + (1.5*iqr)
LIF = q1 - (1.5*iqr)

print('iqr',iqr)
print('UIF',UIF)
print('LIF',LIF)
```

```python
#Check Outlier with Boxplot
dfc_pay.boxplot(column='payment_value', fontsize=10, grid=False, figsize=(8, 8))
```

```python
dfc_cleanpay = dfc_pay[(dfc_pay['payment_value']<=UIF) & (dfc_pay['payment_value']>=LIF)]
dfc_cleanpay
```

```python
dfc_cleanpay.info()
```

# Number of order per-month

```python
order_monthly = df_all[['order_approved_at']].groupby(df_all['order_approved_at'].dt.to_period('M')).agg({'order_approved_at':'count'})
order_monthly.head(20)
```

```python
#Change the column name to num_of_orders
order_monthly = order_monthly.rename(columns = {'order_approved_at': 'num_of_orders'})
#Changing order_approved_at to column, not an index
order_monthly = order_monthly.reset_index()

#adding column for month name and year 
order_monthly['month_year'] = order_monthly['order_approved_at'].dt.strftime('%b-%Y')
order_monthly.head()
```

```python
plt.figure(figsize=(15,8))#ticks
sb.set_theme(style="dark", font_scale=0.7)
order_monthly_viz = sb.lineplot(x = 'month_year', y = 'num_of_orders', data = order_monthly)
order_monthly_viz.set_title('Number of Order per Month 2016-2018',fontsize = 20)
order_monthly_viz.set_xlabel('Month-Year',fontsize = 14)
order_monthly_viz.set_ylabel('Number of Orders',fontsize = 14)
plt.show()
```

```python
def call_monthly_order (df,year):
  df_year = df[df['order_approved_at'].dt.year == year]
  df_year = df_year[['month_year','num_of_orders']]
  sb.set(rc = {'figure.figsize':(10,4)})
  sb.set_theme(style="dark", font_scale=0.8)
  visualize = sb.lineplot(x = 'month_year', y = 'num_of_orders', data = df_year)
  visualize.set_title('Number of Order per Month in '+ str(year), fontsize = 17)
  visualize.set_xlabel('Month',fontsize = 13)
  visualize.set_ylabel('Number of Orders', fontsize = 13)
  return visualize
```

```python
orders_2018 = call_monthly_order(order_monthly,2018)
```

```python
orders_2017 = call_monthly_order(order_monthly,2017)
```

Insight :
From the data above, the highest order was in November 2017 and the lowest order was in December 2016

in 2017 the highest order was in November 
in 2018 the highest order was in January and March

we can see that in 2018 order is pretty fluctuated, start from january until may 

# Daily Order Trend

```python
order_daily = df_all[['order_approved_at']].groupby(df_all['order_approved_at'].dt.dayofweek.sort_index()).count()
order_daily
```

```python
daily_trend_index = order_daily.rename(columns = {'order_approved_at': 'num_of_orders'})
daily_trend = daily_trend_index.reset_index()
daily_trend
```

```python
dow = ['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday']

for i in range(7):
  daily_trend['order_approved_at'] = daily_trend['order_approved_at'].replace(i,dow[i])
daily_trend
```

```python
daily_trend_viz = sb.barplot(x='order_approved_at', y='num_of_orders', data=daily_trend, color = 'blue')
daily_trend_viz.set_xlabel('Number of Order',fontsize = 13)
daily_trend_viz.set_ylabel('Days', fontsize = 10)
plt.title('Number of Order for Each Day',fontsize = 14)
plt.show()
```

INSIGHT 
from the data above, the busiest day is Tuesday and sunday is the lowest day to buy product

# Percentage order per payment type

```python
pay_count_index = df_all[['payment_type']].groupby(df_all['payment_type']).count()
pay_count_index = pay_count_index.rename(columns = {'payment_type':'num_of_payment'})
pay_count = pay_count_index.reset_index()
pay_count
```

```python
label = pay_count['payment_type']
plt.figure(figsize = (18,5))
plt.title('Percentage of Payment Type', fontsize = 15)
plt.pie(pay_count['num_of_payment'], labels = label,autopct = '%0.1f%%')
plt.show()
```

INSIGHT :
Credit card is most preferable payment type, followed by boleto. and and the lowest usage is on a debit card
